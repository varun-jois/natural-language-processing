```python
# Loading the reviews
with open('movie_reviews.pickle', 'rb') as f:
    movie_reviews = pickle.load(f)
```

So movie_reviews is a list of dicts where each dict represents a movie. For example, the first movie is [Goldfinger](https://www.rottentomatoes.com/m/goldfinger). Therefore, we have reviews for 60 movies. The actual reviews along with their ratings can be found in the 'reviews' key in the dict representing each movie.

```python
# Extract the movie reviews and their respective ratings
reviews, ratings = zip(*[(r['review'], r['rating']) for m in movie_reviews for r in m['reviews']])
```
